Visualizing the Distribution of a Categorical or Continuous Variable (using titanic dataset from kaggle)

```python
#importing all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('darkgrid')
```

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['Survived'] = df.Survived.map({0:'Not Survived', 1:'Survived'})
print(df)
```

```python
mapping = {1:'first class', 2:'second class', 3:'third class'}
df['Pclass'] = df.Pclass.map(mapping)
print(df)
```

```python
mapping = {'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}
df['Embarked'] = df.Embarked.map(mapping)
print(df)
```

```python
df.head()
```

```python
df.dtypes
```

Visualization through differnt graphs

```python
#To create a countplot to display the number of male and female passengers on board
ax=sns.countplot(x='Sex',data=df)
plt.title('COUNT OF PASSENGERS AS PER GENDER')
```

From the above plot, we can infer that there are more number of male passengers as compared to the female passengers

```python
# Count the number of passengers in each class
class_counts = df['Pclass'].value_counts()
```

```python
# Create a bar chart of the class counts
plt.bar(class_counts.index, class_counts.values)
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Distribution of Passengers by Class')
plt.show()
```

```python
# Create a histogram of the ages of passengers
plt.hist(df['Age'], bins=10)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Passenger Ages')
plt.show()
```

We can infer that the passenger in the age group of 20-30 years are more in count as per the above visualization

```python
#Creating a coumtplot that displays the count of passengers based on their port of embarked
sns.countplot(x='Embarked',data=df)
plt.title('PORT OF EMBARKED')
plt.xlabel('port')
```

```python
# Create a scatter plot of age vs. fare
plt.scatter(df['Age'], df['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare')
plt.show()
```

From this graph we can infer that the fare is mostly paid by young population

```python
# Create a scatter plot of Pclass vs. fare
plt.scatter(df['Pclass'], df['Fare'])
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Scatter Plot of Pclass vs. Fare')
plt.show()
```

From this graph we can conclude that pasengers who travelled in first class has paid more fare
